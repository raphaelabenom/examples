# Busca em profundidade

```python
def dfs(grafo, inicio, fim, caminho=None):
    print('Temporário', caminho)
    if caminho is None:
        caminho = [inicio]

    if inicio == fim:
        return caminho

    for vizinho in grafo[inicio]: # pilha de execução
        if vizinho not in caminho:
            novo_caminho = dfs(grafo, vizinho, fim, caminho + [vizinho])
            if novo_caminho:
                return novo_caminho
    return None
```

# Busca em largura

```python
from collections import deque

def bfs(grafo, inicio, fim):
    pilha = deque([[inicio]]) # Criar uma pilha inicializada com o nó de partida
    visitados = set() # criar lista vazio para verificação de nós visitados
    while pilha:
        print('Temporário', pilha)
        caminho = deque.popleft(pilha) # extrair caminho mais à esquerda
        no_atual = caminho[-1] # extrair último nó do caminho
        if no_atual not in visitados:
            visitados.add(no_atual)
            if no_atual == fim:
                return caminho
            for vizinho in grafo[no_atual]:
                novo_caminho = list(caminho)
                novo_caminho.append(vizinho)
                pilha.append(novo_caminho)
    return None
```
