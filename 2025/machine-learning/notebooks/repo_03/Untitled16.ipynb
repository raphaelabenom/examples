# Exercício 1

Organização de nomes
Crie uma função chamada organizar_nomes(lista_nomes) que:

- Recebe uma lista de nomes (com duplicatas e maiúsculas/minúsculas misturadas).
- Remove duplicatas.
- Retorna os nomes ordenados em ordem alfabética e padronizados com a primeira letra maiúscula.

```python
'FranciscO'.title()
```

```python
def organizar_nomes(lista_nomes):
    for indice, nome in enumerate(lista_nomes):
        lista_nomes[indice] = nome.title()

    lista_nomes_aux = []
    for nome in lista_nomes:
        inserir = True
        for nome_aux in lista_nomes_aux:
            if nome == nome_aux:
                inserir = False
        if inserir:
            lista_nomes_aux.append(nome)

    lista_nomes = lista_nomes_aux
    lista_nomes.sort()
    return lista_nomes
```

```python
nomes = ['Zebuleu', 'ANA',
         'José', 'Arthur',
         'RAUL', 'SilvIO',
         'Diego', 'maria',
         'Maria', 'MarIA']
organizar_nomes(nomes)
```

# 2. Contagem de palavras em um texto
Escreva uma função contar_palavras(texto) que:
- Recebe uma string.
- Conta quantas vezes cada palavra aparece.
- Retorna um dicionário com a contagem.

```python
def contar_palavras(texto):
    palavras = {}
    texto = texto.lower()
    texto = texto.split()
    for palavra in texto:
        if palavra in palavras:
            palavras[palavra] += 1
        else:
            palavras[palavra] = 1
    return palavras
```

```python
texto = 'O rato roeu a roupa do rei de Roma que era amigo do rato do outro rei'
contar_palavras(texto)
```

# B - 2. Manipulação de Strings e Regex
Escreva uma função extrair_emails(texto) que:
- Recebe o endereço de um arquivo contendo um texto (Arquivo exemplo: ia_1_exemplo_ata_reuniao.txt).
- Usa expressões regulares para encontrar todos os endereços de e-mail válidos.
- Retorna a lista de e-mails encontrados.


```python
import re
def extrair_emails(endereco_texto):
    texto = open(endereco_texto, 'r')
    texto = texto.read()
    emails = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b',
                        texto)
    return emails
```

```python
import requests
```

```python
url_base = 'https://servicodados.ibge.gov.br/api/v2/censos/nomes/'
```

```python
def obter_frequencia(nome):
    r = requests.get(url_base + nome)
    try:
        dados = r.json()
        frequencia_total = 0
        for f in dados[0]['res']:
            frequencia_total += f['frequencia']
    except:
        frequencia_total = 0
    return frequencia_total
```

```python
lista_nomes = ['henricco',
               'rodrigo',
               'geovana',
               'enzo']
for nome in lista_nomes:
    frequencia_total = obter_frequencia(nome)
    print(f'{nome} - {frequencia_total}')
```
